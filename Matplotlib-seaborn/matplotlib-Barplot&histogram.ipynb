```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Sahte veri oluşturma

```python
data = {
    "Yıl" : np.arange(2010,2021),
    "Satış": np.random.randint(100,500,11),
    "Masraf": np.random.randint(50,400,11)
}

df= pd.DataFrame(data)

regions = ["Kuzey","Güney","Doğu","Batı"]
df["Bölge"]= np.random.choice(regions, df.shape[0])

df
```

```python
plt.plot(df["Yıl"], df["Satış"],
         label="Satışlar",
         linestyle = "-",
         marker="o", color="blue",
         markerfacecolor="black")

# Soru: Yıl ve Masrafı aynı plota ekleyin.
# Marker: "x"

plt.plot(df["Yıl"],df["Masraf"],label="maraflar",marker="x",linestyle="-",color="red",markerfacecolor="black")

plt.legend()
plt.show()
```

```python
plt.plot(df["Yıl"], df["Satış"],
         label="Satışlar",
         linestyle = "-",
         marker="o", color="blue",
         markerfacecolor="black")

# Soru: Yıl ve Masrafı aynı plota ekleyin.
# Marker: "x"

plt.plot(df["Yıl"],df["Masraf"],label="maraflar",marker="x",linestyle="-",color="red",markerfacecolor="black")

plt.title("Yıllık Satış ve Masraflar")
plt.xlabel("Yıllar",fontsize=18)
plt.ylabel("TL",fontsize=18)



plt.legend()
plt.show()
```

## plt.bar()
`plt.bar()` bir sütun grafiği çizer. Bu grafik türü, kategorilerin değerleri arasındaki ilişkiyi gösterir.

### Parametreler
`color`: Sütunların rengini belirtir.

`width`: Sütunların genişliğini kontrol eder

`edgecolor`: Sütunların kenarlık rengini belirler

`linewidth`: Kenarlık kalınlığını belirtir.

```python
categories = ["A","B","C","D","E"]
values = [35,20,45,51,10]

plt.bar(categories,values, color="red",
        width=0.5, edgecolor="black", linewidth=2)
```

## Histogram mantığı

```python
sales_bins=pd.cut(df["Satış"], bins=10)
sales_bins  ## Satış değerlerini 10 aralıklı gruba böldük


### İKİ FONKSİYON VAR. "pd.cut" ve "pdqcut"
###  "pd.cut" : bin arası mesafe eksik (tüm aralıklar arasında ki mesafe eşittir. (örn: 10 ile 5 arası, 30 ile 35 arası sayıları arar))

### "pd.qcut" : bins parametresi almaz q paramtersi alır. aralıklar eşit değlldir. random olur.
```

```python
sales_bins.value_counts() ## bize aralıklar arasında hangi değerden kaç tane olduğu yazdırır.
```

```python
sales_bins=pd.qcut(df["Satış"], q=10)
bin_counts=sales_bins.value_counts().sort_index()  ##Bu ise bize verileri indexsırasına göre sıralar.
```

```python
bin_counts.index  ## solve sağ değerleri bulunur örnek 100.999 ve 128.0 
```

```python
## bunları oratlamasını  için ikiye ayırırz ve işlem yaparız .

bin_middles= [(interval.left+interval.right) /2 for interval in bin_counts.index]
bin_middles
```

```python
plt.bar(bin_middles,bin_counts.values,width=15, alpha=0.7 )
```

# hist()

```python
plt.hist(df["Satış"],bins=10,alpha=0.5,label="Satışların frekansı",width=15)  #bins bizim için gruplar ve 100-500 arasında 10 adet bar verir.
plt.legend()
plt.show()
```

### soru: numpy ile random veri oluiturup plt.hist() kullanarak frekansını görselleştirin.
``-np.random.randn kullanabilirsiniz.``

``-bind = 30 olsun.``

``renk vb. farketmez.``

```python
x=np.random.randn(1000)
plt.hist(x,bins=30,width=0.5)
plt.show()
```

```python
x=np.random.randn(1000)
plt.hist(x,bins=30,width=0.5,range=(-10,10))
plt.show()
```

```python
# y ekseni değişti, yoğunluk gösteriyor. normalize değerler 0-1 arasında.

x=np.random.randn(1000)
plt.hist(x,bins=30,width=0.5,range=(-10,10),density=True)
plt.show()
```

## plt.hist()

Nümerik değerlerinizin dağılımını yorumlamanızı sağlayan bir grafiktir, histogram çizer.

### Parametreler
`bins`: Grup sayısını belirler.

`range`: Bin'lerin alt ve üst limitini belirler.

`density`: True olursa grafik yoğunluk belirtir, false olursa frekans belirtir.

`color`: Bin rengi

`alpha`: Saydamlık opaklık
