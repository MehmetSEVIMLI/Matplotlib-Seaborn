# soru : Numpy kullanarak cos(x) ve sin(x) grafilerini çizdirelim.
 * cos(x) plt.scatter(), sin(x) plt.plot
 * x = np.arange(100)/10

```python
x = np.arange(100)/10
y1= np.cos(x)
y2=np.sin(x)

plt.scatter(x,y1, color="black" , label="cosX")
plt.plot(x,y2, label="sinX")
plt.legend()
plt.show()
```

```python
plt.plot(x,y2, label="sinX", color="green", linestyle="--")
plt.legend()
plt.show()
```

## plt.plot ve plt.scatter Parametreleri


### plt.plot
 `color` : çizginin regnini belirler

 `linestyle` : çizgi stilini belirler

 `marker` : noktorlara(kesişimlere) denk gelen kısımmları gösterir. "o" olabilir, "s" olabilir

 `markersize` : marker'ın boutunu belirler

 `alpha` : saydamlık - opaklık. 1=opak, 0=saydam
 
 `label` : grafiğin etiketi

```python
x=np.arange(100)/10
plt.scatter(x,np.cos(x), color="red",label="cosX",s=10)
# s markersize ile aynı işlevi görebilir
plt.legend()
plt.show()
```

```python
x=np.arange(100)/10
plt.scatter(x,np.cos(x), color="red",label="cosX",s=10, edgecolors="blue")
# edgecolor halkaların dışında bir renk vermemizi sağlar
plt.legend()
plt.show()
```

```python
x=np.arange(100)/10
plt.scatter(x,np.cos(x), color="red",label="cosX",s=10, edgecolors="blue", linewidths=10)
plt.legend()
plt.show()
```

```python

x=np.arange(100)/10
plt.scatter(x,np.cos(x), color="red",label="cosX",s=10, edgecolors="blue",linewidths=2,alpha=0.5)
plt.legend()
plt.show()
```

## plt.plot ve plt.scatter Parametreleri

### plt.scatter parametreleri
`color`: renk belirtir

`label`: grafiğin etiketi

`s`: noktaların büyüklüğü

`edgecolors`: noktaları saran renk.

`alpha`: saydamlık ve opaklık
